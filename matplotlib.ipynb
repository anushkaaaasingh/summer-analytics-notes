```python
import matplotlib.pyplot as plt
plt.plot([1,3,2,4])
plt.show()
# here indexes are the x-axis

plt.plot([1,2,3,4],[1,4,9,16])
plt.xlabel('x')
plt.ylabel('x^2')
plt.title('graph of squares', fontsize = 14, color = 'purple')
plt.show()

#Resizing of plot
plt.figure(figsize = (3,5), facecolor="yellow")  # 3 inches wide, 5 inches tall
plt.plot([1,2,3,4],[1,4,9,16], marker = 'o', linestyle="dashed", color="green") # plot appearance
plt.xlabel('x')
plt.ylabel('x^2')
plt.axis([0,6,0,17])
plt.annotate('Square it', (2,8)) # second parameter is the coordinate of annotation
plt.show()
```
